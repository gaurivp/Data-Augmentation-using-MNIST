```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import tensorflow as tf 
from tensorflow import keras 
```

```python
from keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
from matplotlib import pyplot
for n in range(0, 9):
    pyplot.subplot(330 + 1 + n)
    pyplot.imshow(x_train[n], cmap=pyplot.get_cmap('gray'))
    pyplot.show()
    break
```

```python
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1))
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
#Horizontal and vertical flip

datagen1 = ImageDataGenerator(horizontal_flip=True, vertical_flip=True)
datagen1.fit(x_train)

for x_img, y_img in datagen1.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_img[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
#ZCA whitening


datagen2 = ImageDataGenerator(zca_whitening=True)
datagen2.fit(x_train)

for x_img, y_img in datagen2.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_img[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
#Rotation with an angle of 135 degrees


datagen3 = ImageDataGenerator(rotation_range=135)
datagen3.fit(x_train)

for x_img, y_img in datagen3.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_img[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
#Brightness Range


datagen4 = ImageDataGenerator(brightness_range=[0.5,1.5])
datagen4.fit(x_train)


for x_img, y_img in datagen4.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_img[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
#Zoom Range

datagen5 = ImageDataGenerator(zoom_range=[0.2,1.0])
datagen5.fit(x_train)


for x_img, y_img in datagen5.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_img[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
# Shifts

shift = 0.3

datagen6 = ImageDataGenerator(width_shift_range=shift, height_shift_range=shift)
datagen6.fit(x_train)


for x_img, y_img in datagen6.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_img[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
#Random flips


datagen7 = ImageDataGenerator(horizontal_flip=True, vertical_flip=True)
datagen7.fit(x_train)


for x_img, y_img in datagen7.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_img[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```

```python
#Addition of noise 


datagen8 = ImageDataGenerator(horizontal_flip=True, vertical_flip=True)
datagen8.fit(x_train)

noise_factor = 1.0

x_train_with_noise = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape)
x_test_with_noise = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)
x_train_with_noise = np.clip(x_train_with_noise, 0.0, 1.0)
x_test_with_noise = np.clip(x_test_with_noise, 0.0, 1.0)



for x_img, y_batch in datagen8.flow(x_train, y_train, batch_size=9):
	for n in range(0, 9):
		pyplot.subplot(330 + 1 + n)
		pyplot.imshow(x_test_with_noise[n].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
	pyplot.show()
	break
```
